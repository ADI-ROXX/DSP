```python
import numpy as np
import matplotlib.pyplot as plt


fs = 1000
# f = eval(input("Enter the frequency : "))   
f=10
t = np.arange(0, 1, 1/fs)  
x = np.sin(2 * np.pi * f * t)  

# noise_amp = eval(input("Enter the noise amplitude : "))   
noise_amp=0.5
noise = noise_amp * np.random.randn(len(t))
noisy_signal = x + noise
filtlen = 10
fx = np.ones(filtlen) / filtlen
filtered_signal = np.convolve(noisy_signal, fx,mode="same")

plt.figure(figsize=(5,3))
plt.plot(t, x, label='Original Sine Wave')
plt.plot(t, noisy_signal, label='Noisy Signal',alpha=0.6)
plt.plot(t, filtered_signal, label='Filtered Signal',alpha=0.7)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Original vs Noisy vs Filtered')
plt.legend()
plt.grid(True)
plt.show()
```

```python
num_freq=eval(input("Enter the number of frequencies: "))

frequencies = []
for i in range(num_freq):
    frequencies.append(eval(input("Enter the frequency "+str((i+1)))))

combined_signal = np.sum([np.sin(2 * np.pi * freq * t) for freq in frequencies], axis=0)
fourier_transform = np.fft.fft(combined_signal)
frequencies_fft = np.fft.fftfreq(len(combined_signal), 1/fs)

plt.figure(figsize=(5,3))
plt.plot(frequencies_fft, np.abs(fourier_transform))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.xlim(-max(np.abs(frequencies))-10,max(np.abs(frequencies))+10)
plt.title('Fourier Transform of Combined Sine Waves')
plt.grid(True)
plt.show()
```

```python

pulse_duration = eval(input("Enter the duration of the pulse in seconds: "))
pulse = np.where((t >= 0) & (t <= pulse_duration), 1.0, 0.0)
pulse_fft = np.fft.fft(pulse)
frequencies_pulse_fft = np.fft.fftfreq(len(pulse), 1/fs)
plt.figure(figsize=(10,6))
plt.plot(t,pulse)
plt.title("Pulse")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
plt.figure(figsize=(10,6))
plt.plot(frequencies_pulse_fft, np.abs(pulse_fft))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of Pulse Signal')
plt.grid(True)
plt.show()
```
