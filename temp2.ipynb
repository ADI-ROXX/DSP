```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
fs = 1000  # Sampling frequency
T = 7     # Duration of the signal in seconds
freq=1
amp=4

t = np.linspace(0, T, int(fs * T), endpoint=False)
triag = amp * np.abs((t * freq) % 1 - 0.5) * 4

triag_fft = np.fft.fft(triag)
fft_freq = np.fft.fftfreq(len(triag), 1/fs)

# Plot the Fourier Transform
plt.figure(figsize=(10, 6))
plt.plot(fft_freq, np.abs(triag_fft))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of Sampled Triangular Wave')
plt.grid(True)
plt.xlim(-5,5)
plt.show()
```
